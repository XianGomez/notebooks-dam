Ejecuta el contenido del try, si salta error y es el mismo del except corre el contenido del except y del finally
Si el try se ejecutó con éxito, corre la parte del else y del finally
El último bloque del código no se ejecuta y rompe el código porque el except es IndexError, cuando debería ser ZeroDivisionError. Ya que el error ZeroDivisionError es de tipo aritmético

```python
# File mergedexc.py (Python 3.X + 2.X)
sep = '-' * 45 + '\n'

print(sep + 'EXCEPTION RAISED AND CAUGHT')
try:
    x = 'spam'[99]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')

print(sep + 'NO EXCEPTION RAISED')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')

print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
else:
    print('else run')
finally:
    print('finally run')
print('after run')

print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')
try:
    x = 1 / 0
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```
