## Crear y mostrar una cadena de 4 caracteres

```python
# Make a 4-character string, and assign it to a name
S = 'Spam'
print(S)  # muestra la cadena
```

## Longitud y la función `len`

```python
# Obtener longitud
len(S)  # devuelve 4
```

## Indexado (índices base 0)
En Python los índices empiezan en 0: el primer elemento está en la posición 0.

```python
S[0]  # primer caracter -> 'S'
S[1]  # segundo caracter -> 'p'
# Si sales de los límites, obtienes IndexError (ejemplo comentado)
# S[8]  # IndexError: string index out of range
```

## Indexado negativo (desde el final)
Los índices negativos cuentan desde el final: `-1` es el último, `-2` el penúltimo, etc.

```python
S[-1]  # último -> 'm'
S[-2]  # penúltimo -> 'a'
# Equivalente usando len: S[len(S)-1] -> 'm'
S[len(S)-1]
```

## Concatenación y repetición
Usa `+` para concatenar y `*` para repetir.

```python
S + 'eggs'    # 'Spameggs'
S * 3         # 'SpamSpamSpam'
# S permanece sin cambios a menos que lo reasignes
S
```

## Conversión entre strings y listas
Puedes convertir una cadena en una lista de caracteres y viceversa con `list()` y `''.join(...)` para obtener mutabilidad temporal.

```python
E = 'egss'
L = list(E)
print(L)
L[0] = 'z'
L
''.join(L)  # vuelve a cadena: 'zgss'
```

## Métodos útiles de strings (tipo específico)
Ejemplos: `find`, `replace`, `upper`, `isalpha`, `isdigit`, `split`, `rstrip`, `lstrip`, `format` y f-strings.

```python
S = 'SpamEggsSpam'
S.find('pa')        # primer 'pa' -> 1
S.find('pa', 3)     # buscar desde offset 3 -> 9
S.find('z')         # no existe -> -1
S.replace('Eggs','Bacon')  # 'SpamBaconSpam' (S sigue igual)
S.upper()
print('S is alpha', S.isalpha())
print('S is digit', S.isdigit())
```
