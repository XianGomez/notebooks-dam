```python
L = [2, 3, 4]
M = L
L[0] = 24
print("L = ", L)
print("M = ", M)
```

L guarda una lista en memoria, y M APUNTA A L, se modifica el primer int de la lista, por lo que tambien se cambia M, ya que no crea una copia, solo se iguala a L

```python
L = [2, 3, 4]
M = L[:]
L[0] = 24

print("L = ", L)
print("M = ", M)
```

L se iguala a una lista, M se iguala al contenido de la lista de L copiandolo, L cambia la posicion 0 por 24

```python
L = [1, 2, 3]
M = L

L == M
L is M
```

L es igual a una lista, M = L y después nos L al ser totalmente igual a M nos sale True y en el L is M nos sale true porque M se igualó a L

```python
L = [1, 2, 3]
M = [1, 2, 3]

print(
    "L == M",
    L == M 
)

print(
    "L is M",
    L is M
)
```

Los int de las listas L y M son iguales entonces al poner L == M esto va a dar True, pero si pones que L es M te pondrá False ya que son listas diferentes
