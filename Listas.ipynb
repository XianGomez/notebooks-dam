#### Iteracion y Comprension

```python
#Confirmar si hay un item en la lista
3 in [1,2,3]
```

```python
"white" in ["white", "blue", "yellow"]
```

```python
#En cada iteracion la variable x coge el valor del elemento actual
for x in [1,2,3]:
    print(x, end=" ")
#end="" despues de cada elemento printado pone lo que está entre ""
#Si no pone por defecto un salto de linea
```

```python
#Caracter por caracter en SPAM, separa el string en caracteres
res = [c for c in "SPAM"]
res
```

```python
#Lo mismo que lo de arriba pero de otra forma
res2 = [] #Crea la lista
for c in "SPAM": #Por cada caracter en SPAM
    res2.append(c) #Añade valor al final de la lista
print(res2) #Printa la lista
```

```python
#map() aplica una funcion a cada elemento de la lista, en este caso la funcion abs()
list(map(abs, [-1, -2, -3]))
```

#### Indexing, Slicing and Matrixes

```python
#Acceso a un item de la lista, empieza en 0
L = ["yellow", "green", "blue", "red"]
L[0]
```

```python
#Acceso al ultimo item de la lista -1
L[-1]
```

```python
#Acceso a los items desde la posicion ingresada incluyendolo
L[1:]
```

```python
#Acceso a los items hasta la posicion ingresada sin incluirlo
L[:-1]
```

```python
#Lista de listas, Acceso a los items atraves de fila y columna
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
matrix[1][1]
#matrix[fila][columna]
```

#### Cambiar Items

```python
M = ["Spam", "spam", "SPAM"]
M[1] = "SpAm"
M
```

```python
#Reemplaza la lista por rango de posiciones
M[0:2] = ["Eat", "more"]
M
```

```python
#Reemplazar
M[0:2] = ["Dont", "Eat"]
M
```

```python
#Añadir /Poner la misma posición para añadir/
M[0:0] = ["DONT", "EAT"] 
M
```

```python
#Borrar
M[0:2] = []
M
```

```python
#Añadir al principio
M[:0] = ["EGGS","and"]
M
```

```python
#Añadir al final, len devuelve la longitud de la lista, si la lista es larga nos sirve
M[len(M):] = ["E"]
M
```
