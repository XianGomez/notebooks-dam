```python
def changer(a, b):
		a = 2
		b[0] = 's'
					
x = 1
l = [1, 2]
changer(x, l)
print("l = ", l)
assert x == 1, "x es 1"
```

la funcion changer hace que x cambie su valor por a y l cambie su primer valor por b ya que b[0] = "s" y al printar l sale el cambio de b y la posicion 1 de la lista

```python
l = [1, 2]
changer(x, l[:])

print("l = ", l) 
```

est치 pasando una copia de l, por lo que se cambia la copia de l, la original queda intacta


```python
l = [1, 2]
def changer(a, b):
    b = b[:]		
    a = 2
    b[0] = 's'
    print(a)

changer(x, l)
assert l == [1, 2], "l es una invariante"
print("l = ", l) 
```

El changer solo cambia el valor de las variables entonces al cambiar l por b, se cambia el valor no el nombre de la variable, entonces seria como tener b = l[:]

```python
l = [1, 2]
def changer(a, b):	   
		a = 2			
		b[0] = 's'

changer(x, tuple(l))

print("l = ", l)
assert l == [1, 2], "l es una invariante"
```

Diferentes formas de dar valor a los parametros

```python
def func(spam, eggs, toast=0, ham=0):
	print((spam, eggs, toast, ham))

func(1, 2)
func(1, ham=1, eggs=0)
func(spam=1, eggs=0)
func(toast=1, eggs=2, spam=3)
func(1, 2, 3, 4)
```

Diferentes formas de darle valor a los paramentros


```python
def kwonly(a, *b, c):
    print(a, b, c)

kwonly(1, 2, c=3)

kwonly(a=1, c=3)

kwonly(1, 2, 3)
```

el asterisco hace que todos los valores apartir de donde se ponga se meta en la variable b como tupla

```python
def kwonly(a, *, b='spam', c='ham'):
	print(a, b, c)

kwonly(1)
kwonly(1, c=3)
kwonly(a=1)
kwonly(c=3, b=2, a=1)
kwonly(1, 2)  
```
