# Operadores lógicos y bit a bit en Python

## Introducción
En este cuaderno se muestran ejemplos prácticos del uso de operadores lógicos en Python.

## Operador `and`

```
counter = 1
value = 100
counter > 0 and value == 100
```

```
def demo_and(a, b):
    if not a:
        return a
    return b
demo_and(True, False), demo_and(False, True)
```

## Operador `or`

```
def demo_or(a, b):
    if a:
        return a
    return b
demo_or(True, False), demo_or(False, True)
```

## Operador `not`

```
not True, not False, not [], not [1]
```

## Expresiones equivalentes

```
var = 1
print(var > 0)
print(not (var <= 0))
print(var != 0)
print(not (var == 0))
```

## Cadena de comparaciones

```
x = 3.141
if 3.14 < x < 3.142:
    print('x está cerca de pi')
```

## Precedencia de operadores

```
1 * 4 < 5 + 1 and True != 0
```

## Valores lógicos y bits

```
not not 1, not not 0
```
